### impute missing values for dropoffs and lbs column


**linear regression on non-missing values to estimate dropoffs from lbs or vise versa**


```python
temp_df = df[['dropoffs','lbs']].copy()
```

```python
temp_df = temp_df.dropna()
```

```python
from sklearn.linear_model import LinearRegression

# Split the data into independent and dependent variables
X = temp_df[['dropoffs']]
y = temp_df['lbs']

# Create an instance of the LinearRegression model
model = LinearRegression(fit_intercept=False)

# Fit the model to the data
model.fit(X, y)

# Print the coefficient and intercept
print("Coefficient:", model.coef_)
```

### csv of daily historical weather data downloaded from open-meteo.com

```python

orig_df = pd.read_csv("data/historical_weather.csv")
weather_df = orig_df.copy()
```

```python
weather_df.head()
```

### cleaning weather dataframe

```python
new_column_names = {'latitude': 'date',
                    'longitude': 'temp',
                    'elevation': 'rain',
                    'utc_offset_seconds': 'snow',
                   }

weather_df = weather_df.rename(columns=new_column_names)
weather_df = weather_df.iloc[2:]
weather_df = weather_df.drop(columns=['timezone','timezone_abbreviation'])
weather_df = weather_df.reset_index(drop=True)

weather_df.head()
```

**finding date when Sunday collection started (from composting df):**

```python
# Find the first instance of a value
index = df['sunday_dropoffs'].eq(True).idxmax()

df.iloc[index] # index 69, December 12th 2021
```

**finding sunday collection date (Dec 12th 2021) in weather_df:**

```python
index = weather_df['date'].eq('2021-12-12').idxmax()

weather_df.iloc[index] # index 64
```

**splitting weather dataframe into seperate dataframes for Saturday pickup and Sunday pickup dates:**

```python
sat_pickup_weather_df = weather_df.iloc[:64]
sun_pickup_weather_df = weather_df.iloc[64:]

sat_pickup_weather_df
```

```python
sun_pickup_weather_df
```

**keeping only Saturdays for Saturday pickup dataframe:**

```python
# keeping only saturdays in sat_pickup_weather_df
sat_pickup_weather_df = sat_pickup_weather_df[sat_pickup_weather_df.index % 7 == 0]
sat_pickup_weather_df = sat_pickup_weather_df.iloc[:-1]
sat_pickup_weather_df
```

**keeping only Sundays for Sunday pickup dataframe:**

```python
# keeping only saturdays in sat_pickup_weather_df
sun_pickup_weather_df = sun_pickup_weather_df.reset_index(drop=True)
sun_pickup_weather_df = sun_pickup_weather_df[sun_pickup_weather_df.index % 7 == 0]
sun_pickup_weather_df
```

**stacking Saturday pickups and Sunday pickups back into one dataframe:**

```python
# stacking sat and sun pickup weather data

sat_and_sun_weather_df = pd.concat([sat_pickup_weather_df,sun_pickup_weather_df], ignore_index=True)
```

```python
# changing date column type to 'datetime64[ns]' and rest of columns to numeric

sat_and_sun_weather_df['date'] = pd.to_datetime(sat_and_sun_weather_df['date'])
sat_and_sun_weather_df['temp'] = pd.to_numeric(sat_and_sun_weather_df['temp'])
sat_and_sun_weather_df['rain'] = pd.to_numeric(sat_and_sun_weather_df['rain'])
sat_and_sun_weather_df['snow'] = pd.to_numeric(sat_and_sun_weather_df['snow'])
sat_and_sun_weather_df.info()
```

```python
# creating boolean columns for significant rain/snow
```

```python
sat_and_sun_weather_df['significant_rain'] = sat_and_sun_weather_df['rain'] >= .33
sat_and_sun_weather_df['significant_snow'] = sat_and_sun_weather_df['snow'] >= .33

sat_and_sun_weather_df.info()
```

***merging weather dataframe with compost dataframe, starting at the right date (Oct 9 2021, when Greenmarket also started collecting scraps)**

```python

complete_df = pd.merge(df, sat_and_sun_weather_df, on='date', how='left')
complete_df.tail()
```
